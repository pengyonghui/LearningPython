```python
# matrix_vectore_product.py

from numpy import zeros, mat, transpose
x = zeros(2)
x = mat(x)
x = transpose(x)
x[0] = 3
x[1] = 2   # Pick some values
A = zeros((2,2))
A = mat(A)
A[0,0] = 1;  
A[0,1] = 0
A[1,0] = 0; 
A[1,1] = 1
# The following gives y = x since A = I, the identity matrix
y = A*x
print (y)
```

```python
#ball_max_height.py
import matplotlib.pyplot as plt

v0 = 5 # Initial velocity
g = 9.81 # Acceleration of gravity
t = linspace(0, 1, 1000)  # 1000 points in time interval
y = v0*t - 0.5*g*t**2     # Generate all heights

# At this point, the array y with all the heights is ready.
# Now we need to find the largest value within y.

largest_height = y[0]          # Starting value for search
for i in range(1, 1000):
    if y[i] > largest_height:
        largest_height = y[i]

print ("The largest height achieved was %f m" % (largest_height))
# We might also like to plot the path again just to compare
plt.plot(t,y)
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.show()
```

### List and tuples

List:
    (1) the number of elements in a list is allowed to change
    (2) Elements in a list can be of different type, i.e you can mix integers, floats and strings
    (3) Lists provide more flexibility than do arrays  
    (4) uses "[]"    
    
Array:
    (1) Arrays have a fixed length that must be known at the time of memory allocation
    (2) Must be of the same type
    (3) Faster computations than list
    (4) require less memory use
    (5) Vectorization requires arrays to be used.

#### List comprehension

```python
List_1 = [1, 2, 3, 4]
# run through all elements of a list, do the same operation on each, and store the new
# elements in another list
List_2 = [e*10 for e in List_1]
```

#### zip()
To run through 2 or more lists at the same time

Tuples:
    (1) very much like lists, but cannot be changed "constant list"
    can't use insert or append
    (2) Quicker for python to handle data in a tuple than in a list
    (3) use "()"
